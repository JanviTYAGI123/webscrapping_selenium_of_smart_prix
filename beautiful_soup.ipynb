```python
with open("smartprix.html", "r",encoding='utf-8') as f:
    html = f.read()
    
```

```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
soup=BeautifulSoup(html, 'lxml')
soup.prettify()
```

```python
containers=soup.find_all('div',{'class':'sm-product has-tag has-features has-actions'})
```

```python
model = []
price = []
ratings = []
sim = []
ram = []
processor = []
display = []
camera = []
card = []
os = []

for i in containers:
    # Name
    try:
        model.append(i.find('h2').text.strip())
    except:
        model.append(np.nan)

    # Price
    try:
        price.append(i.find('span', {'class': 'price'}).text.strip())
    except:
        price.append(np.nan)

    # Rating
    try:
        score_div = None
        for rank in ['rank-1-bg', 'rank-2-bg', 'rank-3-bg']:
            score_div = i.find('div', {'class': f'score {rank}'})
            if score_div:
                break
        ratings.append(score_div.find('b').text.strip() if score_div and score_div.find('b') else np.nan)
    except:
        ratings.append(np.nan)

    # Features list
    try:
        x = i.find('ul', {'class': 'sm-feat specs'}).find_all('li')
    except:
        x = []

    # Sim
    try:
        sim.append(x[0].text.strip())
    except:
        sim.append(np.nan)

    # Processor
    try:
        processor.append(x[1].text.strip())
    except:
        processor.append(np.nan)

    # RAM
    try:
        ram.append(x[2].text.strip())
    except:
        ram.append(np.nan)

    # Display
    try:
        display.append(x[4].text.strip())
    except:
        display.append(np.nan)

    # Camera
    try:
        camera.append(x[5].text.strip())
    except:
        camera.append(np.nan)

    # Card slot
    try:
        card.append(x[6].text.strip())
    except:
        card.append(np.nan)

    # OS
    try:
        os.append(x[7].text.strip())
    except:
        os.append(np.nan)
```
