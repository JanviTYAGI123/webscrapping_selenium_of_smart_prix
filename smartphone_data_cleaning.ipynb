```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('mobile_data.csv')
```

```python
#cleaning the price column
df['Price'] = df['Price'].str.replace('₹','').str.replace(',','').astype(int)
```

```python
df
```

```python
#rating column
```

```python
df['Rating'].fillna(df['Rating'].mean(), inplace=True)
```

```python
df.info()
```

```python
# handling the sim column (by splitting them as :SIM_Type	,Supports_3G,	Supports_4G,	Supports_5G,	VoLTE	WiFi	,NFC	IR_Blaster)
# SIM Type
df['SIM_Type'] = df['SIM'].str.extract(r'(Single Sim|Dual Sim)')

# Network support
df['Supports_3G'] = df['SIM'].str.contains('3G').astype(int)
df['Supports_4G'] = df['SIM'].str.contains('4G').astype(int)
df['Supports_5G'] = df['SIM'].str.contains('5G').astype(int)

# Extra features
df['VoLTE'] = df['SIM'].str.contains('VoLTE').astype(int)
df['WiFi'] = df['SIM'].str.contains('Wi-Fi').astype(int)
df['NFC'] = df['SIM'].str.contains('NFC').astype(int)
df['IR_Blaster'] = df['SIM'].str.contains('IR Blaster').astype(int)
```

```python
df.head()
```

```python
#handling the ram column as also spiltting them  as  ram_gb and storage_gb
# Extract RAM (before the word RAM)
df['RAM_GB'] = df['RAM'].str.extract(r'(\d+)\s*GB RAM').astype(float).fillna(0).astype(int)

# Extract Storage (after the comma)
df['Storage_GB'] = df['RAM'].str.extract(r',\s*(\d+)\s*GB').astype(float).fillna(0).astype(int)
```

```python
df.head()
```

```python
#handling the Processor column  by splitting into  chipset ,cores , clockspeed_ghz
# Extract Chipset name (before first comma)
df['Chipset'] = df['Processor'].str.split(',').str[0]

# Extract number of cores
df['Cores'] = df['Processor'].str.extract(r'(\d+)\s*Core').astype(float)

# Extract clock speed in GHz
df['ClockSpeed_GHz'] = df['Processor'].str.extract(r'([\d\.]+)\s*GHz').astype(float)
```

```python
df.sample(5)
```

```python
df.info()
```

handling the core  element


```python
# Define a mapping
core_map = {
    'Single': 1,
    'Dual': 2,
    'Triple': 3,
    'Quad': 4,
    'Penta': 5,
    'Hexa': 6,
    'Octa': 8,
    'Deca': 10
}

# Extract core word (like Octa, Quad, etc.)
df['Core_Word'] = df['Processor'].str.extract(r'(\w+)\s*Core')

# Map to numbers
df['Cores'] = df['Core_Word'].map(core_map)
```

```python
df['Cores']
```

```python
#handling the ddisplay column
# Extract display size in inches
df['Display_Size'] = df['Display'].str.extract(r'([\d\.]+)\s*inches').astype(float)

# Extract resolution
df['Resolution'] = df['Display'].str.extract(r'(\d{3,4}\s*x\s*\d{3,4})')

# Extract refresh rate in Hz
df['RefreshRate_Hz'] = df['Display'].str.extract(r'(\d+)\s*Hz').astype(float)
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
#handling the camera column
# Rear camera → pick the first MP number before "&"
df['Rear_Main_MP'] = df['Camera'].str.extract(r'(\d+)\s*MP')

# Front camera → MP number after "&"
df['Front_MP'] = df['Camera'].str.extract(r'&\s*(\d+)\s*MP')

# Convert to numeric
df['Rear_Main_MP'] = df['Rear_Main_MP'].astype(float)
df['Front_MP'] = df['Front_MP'].astype(float)
```

```python
df.head()
```

```python
df['Resolution']
```

```python


# Split into Width and Height
df[['Resolution_Width', 'Resolution_Height']] = df['Resolution'].str.replace('×','x').str.split('x', expand=True)

# Convert to numeric
df['Resolution_Width'] = df['Resolution_Width'].str.strip().astype(float)
df['Resolution_Height'] = df['Resolution_Height'].str.strip().astype(float)

# Check result
df[['Resolution','Resolution_Width','Resolution_Height']].head()
```

```python
df.head()
```

```python
#handling card column
```

```python
#handling card column
# Card support yes/no
df['Card_Support'] = df['Card'].fillna('').apply(lambda x: 'No' if 'Not Supported' in x else 'Yes')

# Extract capacity in GB (very simple)
df['Card_Capacity_GB'] = df['Card'].str.extract(r'(\d+)').fillna(0).astype(int)

# Drop old Card column
df.drop(columns=['Card'], inplace=True)

# Preview
df[['model', 'Card_Support', 'Card_Capacity_GB']].head()
```

```python
df.info()
```

```python
#handling os column
# Step 1: Extract OS Name (Android, iOS, etc.)
df['OS_Name'] = df['OS'].str.extract(r'([A-Za-z]+)')

# Step 2: Extract OS Version (like 15, 14, etc.)
df['OS_Version'] = df['OS'].str.extract(r'(\d+)').astype(float)

# Step 4: Preview
print(df[['model', 'OS_Name', 'OS_Version']].head())
```

```python
#filling up  the os version
# Fill missing OS version with most frequent one
df['OS_Version'].fillna(df['OS_Version'].mode()[0], inplace=True)
```
