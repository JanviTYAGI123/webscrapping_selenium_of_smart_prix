```python
import pandas as pd
import numpy as np

# Fill missing values
df.fillna({'OS': 'Unknown', 'OS_Name': 'Unknown', 'SIM_Type': 'Unknown',
           'Chipset': 'Unknown', 'Core_Word': 'Unknown', 'Camera': 'Unknown'}, inplace=True)

df.fillna(0, inplace=True)   # for numeric columns

# Extract brand from model
df['Brand'] = df['model'].str.split().str[0]

# Drop duplicates
df.drop_duplicates(subset=['model'], inplace=True)

# Handle outliers in Price (cap at 99th percentile)
q99 = df['Price'].quantile(0.99)
df['Price'] = np.where(df['Price'] > q99, q99, df['Price'])
```

Right skew (positive skew):

Most phones in your dataset are cheap to mid-range.

A few phones are very expensive (flagships/foldables), which “pull” the tail of the distribution to the right.

Example: ₹8,000 – ₹20,000 phones are common, but some phones cost ₹80,000+ → that creates the skew.

knn inputer

```python
from sklearn.impute import KNNImputer
# select only numeric columns (KNNImputer works on numbers)
num_df = df.select_dtypes(include=['float64', 'int64'])

# create KNN imputer
imputer = KNNImputer(n_neighbors=5)

# fit + transform
num_df_imputed = imputer.fit_transform(num_df)

# put back into DataFrame
df[num_df.columns] = num_df_imputed
```
